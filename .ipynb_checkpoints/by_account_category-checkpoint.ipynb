```python
import pandas as pd
import re
import matplotlib.pyplot as plt

data = "../data/"#data = "/IRA je sais pas quoi
```

```python
df = pd.read_csv(data + "IRAhandle_tweets_1.csv")
for i in range(2, 10):
    df = df.append(pd.read_csv(data + "IRAhandle_tweets_{}.csv".format(i)), ignore_index=True)
    
df = df.dropna()
```

```python
hashtags = []
for t in df.content.values:
    m = re.search("#.* ", t)
    if m is not None:
        for hashtag in m.group(0).split(" "):
            if hashtag and hashtag[0] == "#":
                hashtags.append(hashtag)
        
```

```python
import collections

a = collections.Counter(hashtags)

df["publish_date"] =pd.to_datetime(df["publish_date"], format='%m/%d/%Y %H:%M')
#print(df["publish_date"])
df["harvested_date"] =pd.to_datetime(df["harvested_date"], format='%m/%d/%Y %H:%M')

df
```

```python
def related(df, hashtag, category=None):
    dump = ["", "the", "-", "of", "is", "to", "a", "in", "for", "and", "The", "on", "you", "our", "not", "&", "with", \
           "We", "are", "I", "be", "with", "we", "your", "To", "will", "who", "In", "via", "that", "In", \
           "Is", "this", "it", "by", "THE", "their", "my", "1.", "2.", "all", "at", "Get", "get", "up", \
           "out", "new", "some", "about", "This", "have", "was", "as", "from", "they", "has", "his", "You"\
           "what", "he", "an", "but", "You", "what", "so", "if", "more", "do", "like", "just", "can", "how", \
           "If", "A", "or", "should", "For", "no", "Of", 'one', 'With', "It's", 'He', "I'm", 'want', 'And', 'They', 'when',\
           'It', 'My', 'would', 'US', '4', 'back', 'What', 'us', 'New', 'going', 'Great', 'over', 'time', "don't", 'after', \
            'know', 'than', 'think', 'need', 'take', 'On', "because", "me", "Who", 'ÔøΩÔøΩ','ÔøΩ']
    hashtags = []
    people = []
    words = []

    if category is not None:
        df = df[df.account_category == category]
        
    tweets = [x for x in df.content.values if hashtag in x]
    for tweet in tweets: 
        for word in tweet.split(" "):
            if word and word not in dump:
                if word[0] == "#":
                    hashtags.append(word)
                elif word[0] == "@":
                    people.append(word)
                else:
                    words.append(word)
    
    all_ = [collections.Counter(hashtags), collections.Counter(people), collections.Counter(words)]
    for counter in all_:
        print([a for (a, _) in counter.most_common(30)])
        print()
    return tuple(all_)
```

```python
hashtags, people, words = related(df, "#GiftIdeasForPoliticians")
```

```python
def histograms_bydate(df):
    d= df["publish_date"].groupby(df["publish_date"].dt.date).apply(pd.value_counts)
    return d
def histograms_by_account_type(df):
    d = df[["publish_date", "account_category"]].groupby([df["publish_date"].dt.date, df["account_category"]]).agg("count")
    return d
```

```python
res=histograms_bydate(df)
res= res.groupby("publish_date").sum()
#res
```

```python
ax =res.plot()
plt.show()
```

```python
by_account= histograms_by_account_type(df)
by_account=by_account.drop("publish_date", axis=1)
by_account.columns = ["count"]
by_account

```

```python
by_account = by_account.reset_index().set_index("publish_date")
for cat in by_account.account_category.unique():
    tmp_df = by_account[by_account.account_category == cat]
    ax = tmp_df.plot(title=cat)

plt.show()
```
